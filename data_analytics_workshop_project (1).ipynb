```python
data=clean_data.dropna()
data
```

```python
clean_data["Attendance"]=clean_data["Attendance"].fillna(clean_data["Attendance"].median())
```

```python
clean_data["Gender"]=clean_data["Gender"].fillna(clean_data["Gender"].mode()[0])
```

```python
clean_data["Study_Hours"]=clean_data["Study_Hours"].fillna(clean_data["Study_Hours"].mean())
```

```python
clean_data.duplicated().sum()
```

```python
clean_data[clean_data.duplicated()]
```

```python
clean_data=clean_data.drop_duplicates()
clean_data
```

```python
clean_data["Final_Exam_Marks"]=clean_data["Final_Exam_Marks"].astype(float)
```

```python
clean_data["Gender"]=clean_data["Gender"].str.strip()
```

```python
clean_data.isnull().sum().sum()
```

```python
data.columns=data.columns.str.replace("_"," ")
data
```

```python
data.columns=data.columns.str.replace(" ","_")
data
```

```python
clean_data["Age"]=clean_data["Age"].astype(int)
data
```

```python
clean_data["Percentage"]=clean_data["Percentage"].astype(float)
data
```

```python
clean_data["Total Score"]=(clean_data["Assignment_Score"])+(clean_data["Internal_Marks"])+(clean_data["Final_Exam_Marks"])
clean_data
```

```python
clean_data["External_Marks"]=clean_data["Internal_Marks"]+(clean_data["Assignment_Score"])-(clean_data["Final_Exam_Marks"])
clean_data
```

```python
clean_data["Total_Percentage"]=(clean_data["Total_Marks"]/100)*100
clean_data
```

```python
def grade(Total_Marks):
  if Total_Marks>=90:
    return "A"
  elif Total_Marks>=80:
    return "B"
  elif Total_Marks>=70:
    return "C"
  elif Total_Marks>=60:
    return "D"
  else:
    return "F"
clean_data["Grade++"]=clean_data["Percentage"].apply(grade)
clean_data
```
